```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def create_timeseries(seq_len):
    freq = (np.random.random()*0.5) +1
    amp1 = np.random.random()+0.5
    noise = [np.random.random()*0.3 for i in range(seq_len)]
    x = np.sin(np.arange(0,seq_len)*freq)*amp1+noise
    return (x)
```

```python
plt.plot(create_timeseries(100));
```
